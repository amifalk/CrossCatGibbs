```python
from crosscat import CrossCat
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng
```

```python
top = default_rng().normal(loc=0, scale=1, size=(5,1))
middle = default_rng().normal(loc=10, scale=1, size=(5,1))
bottom = default_rng().normal(loc=-10, scale=1, size=(5,1))

data = np.concatenate((top, middle, bottom), axis=0)

right_top = default_rng().normal(loc=30, scale=1, size=(7,1))
right_bottom = default_rng().normal(loc=-30, scale=1, size=(8,1))
right = np.concatenate((right_top, right_bottom), axis=0)
data = np.concatenate((data, right), axis=1)
```

```python
sns.heatmap(data)
```

```python
model = CrossCat(data, n_chains=4)
```

```python
model.sample(100)
```

```python
model.save("3-2.pkl")
```

```python
for state in model.best_fits():
    state.plot()
```
